```python
f = h5py.File(name='/'.join([filedir,filename]))
```

```python
rys = []
fwhmsx = []
fwhmsy = []
photocoles = []

for name in f:
    metadata = f[name].attrs
    img2 = f[name]['dvf2']
    projx = np.sum(img2,axis=0)
    projy = np.sum(img2,axis=1)

    rys.append(metadata['ry_pos'])
    fwhmsx.append(fwhm(projx))
    fwhmsy.append(fwhm(projy))
    photocoles.append(metadata['photocollector'])
```

```python
slic = slice(4,-4)

fig, (axx, axy) = plt.subplots(1, 2, figsize=(12, 6))

axy.plot(rys[slic], fwhmsy[slic])
axy.grid()
axx.grid()
# plt.xlim(0.35,0.353)
axx.plot(rys[slic], fwhmsx[slic])
plt.show()
```
